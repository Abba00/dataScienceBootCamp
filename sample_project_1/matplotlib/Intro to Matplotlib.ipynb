# Matplotlib example workflow

```python
# import matlplotlib and get it ready for Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize= (10, 10))  #size of 10 x 10 (width and height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Subplot options

```python
# Option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# Plot to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.randn(10), np.random.randn(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2 

fig, ax = plt.subplots(nrows=2,
                       ncols=3,
                       figsize=(20, 5))

ax[0,1].plot(x, x/2);
ax[1,2].scatter(x, np.exp(x));
ax[1,1].bar(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
ax[0,0].hist(np.random.randn(100));
```

### Pyplot vs Matplotlib OO Method

Use Pyplot for quick plots
Use Matplotlib OO method for more advanced projects

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
## OO Method

fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);

ax.set_xlim([45, 80])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

# Customize Plot
ax.set(title="Heart Disease & Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#Add a Legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach

#the fig variable is like the whole space where the graphs(ax) are going to be placed. 

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])

#Customize ax0
ax0.set(title="Heart Disease & Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50['target'])

#Customize ax1
ax1.set(title="Heart Disease & Max Heart Rate",
        ylabel="Thalach")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')


# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle = '--');

# Add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
# Set the style
plt.style.use("seaborn-whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"], #c is for color indexing
                     cmap = "winter") #cmap changes the color scheme

# Customize Plot
ax.set(title="Heart Disease & Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#Add a Legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
# Customizing the y and x axis limitations

# Subplots of chol, age, thalach

#the fig variable is like the whole space where the graphs(ax) are going to be placed. 

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap="winter")

#Customize ax0
ax0.set(title="Heart Disease & Cholesterol Levels",
        ylabel="Cholesterol")
ax0.set_xlim([50, 80])  #Change x axis limits
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50['target'],
                      cmap="winter")

#Customize ax1
ax1.set(title="Heart Disease & Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')


# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle = '--');

# Add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
